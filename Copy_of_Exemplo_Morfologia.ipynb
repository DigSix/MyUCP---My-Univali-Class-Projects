```python
#My code --> Diogo Viana

import cv2
import numpy as np

from google.colab.patches import cv2_imshow

#Ignore
def line_jumper(lines):
  for l in range(lines):
    print("\n")


#All morphological operations need two inputs: the image and a kernel (something like the "instructions" to be followed by the operation).
#1° Units Images
#Here is where we read the images using the function from OpenCV: cv2.imread("Path of file", color_scale) --> Note: "0" is a simplified mode of cv2.IMREAD_GRAYSCALE.
img = cv2.imread('j.png',0)
img_o = cv2.imread('j_ruido.png',0)
img_c = cv2.imread('j_furos.png',0)

#2° Units Kernel
#Here is where we define the kernel using the function from NumPy: np.ones((matrix_i, matrix_j),...).
kernel = np.ones((5,5),np.uint8)
kernel2 = np.ones((15,15),np.uint8)

#Operations

erosion = cv2.erode(img,kernel,iterations = 5)

#This piece of code it's response by calculate the dilation of image
dilation = cv2.dilate(img,kernel,iterations = 5)
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
opening = cv2.morphologyEx(img_o, cv2.MORPH_OPEN, kernel2)
closing = cv2.morphologyEx(img_c, cv2.MORPH_CLOSE, kernel2)

#Results
#Erosion and Dilation
print("Image Erosion")
cv2_imshow(erosion)
print("\nNormal Image")
cv2_imshow(img)
print("\nImage Dilation")
cv2_imshow(dilation)
line_jumper(5)

#Opening
print("Normal Image")
cv2_imshow(img_o)
print("\nOpening Image")
cv2_imshow(opening)
line_jumper(5)

#Closing
print("Normal Image")
cv2_imshow(img_c)
print("\nClosing Image")
cv2_imshow(closing)
line_jumper(5)

#Gradient
print("Normal Image")
cv2_imshow(img)
print("\nImage Gradient")
cv2_imshow(gradient)
line_jumper(5)
```
